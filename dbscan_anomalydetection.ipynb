```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
x,y = make_circles(n_samples=750,factor=.3,noise=0.1)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.15)
x_labels = dbscan.fit_predict(x)
target = np.where(x_labels<0)
```

```python
plt.scatter(x[:,0],x[:,1], c=x_labels)
```

```python
plt.scatter(x[target,0],x[target,1])
plt.title("Anomalies detected by DBSCAN")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
